```python
import sqlite3

conn = sqlite3.connect("keio.db")
cur = conn.cursor()

# Stationテーブル作成
cur.execute("""
CREATE TABLE IF NOT EXISTS Station (
  station_id INTEGER PRIMARY KEY AUTOINCREMENT,
  station_name TEXT,
  limited_express INTEGER
)
""")

# TravelTimeテーブル作成
cur.execute("""
CREATE TABLE IF NOT EXISTS TravelTime (
  time_id INTEGER PRIMARY KEY AUTOINCREMENT,
  station_id INTEGER,
  minutes INTEGER,
  FOREIGN KEY(station_id) REFERENCES Station(station_id)
)
""")

conn.commit()
```

```python
cur.execute("SELECT COUNT(*) FROM Station;")
print(cur.fetchone())

cur.execute("SELECT COUNT(*) FROM TravelTime;")
print(cur.fetchone())
```

```python
import sqlite3
import pandas as pd

# ① DBに接続
conn = sqlite3.connect("keio.db")

# ② SQLでデータ取得
sql = """
SELECT
  s.limited_express,
  t.minutes
FROM Station s
JOIN TravelTime t
ON s.station_id = t.station_id
"""

# ③ DataFrameにする
df = pd.read_sql(sql, conn)

print(df.head())
```
